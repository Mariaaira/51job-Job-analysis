```python
# 导入库
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_csv(r'D:\Crawler\51job\数据最最最新(1).csv',  encoding='gb18030')
data.dropna(inplace=True, axis=0)
data['月薪'] = (data['月薪最大值']+data['月薪最小值'])/2
# 选择特征
features = ['月薪', '工作经验要求', '学历要求']

# 数据标准化
for feature in features:
    data[feature] = (data[feature] - data[feature].mean()) / data[feature].std()

# K-Means聚类分析
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(data[features])

# 分析聚类结果
labels = kmeans.labels_
data['label'] = labels

# 可视化聚类结果
colors = ['red', 'green', 'blue', 'yellow', 'orange']
for i in range(len(colors)):
    cluster = data[data['label'] == i]
    plt.scatter(cluster['月薪'], cluster['工作经验要求'], c=colors[i], alpha=0.5)
plt.xlabel('Salary')
plt.ylabel('Work Experience')
plt.show()
```
